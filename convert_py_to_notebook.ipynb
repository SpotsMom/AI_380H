# How to Convert a Python Script (.py) into a Jupyter Notebook
This notebook demonstrates the steps to convert a Python script into a Jupyter notebook using Visual Studio Code.

## 1. Open the .py File in Visual Studio Code
Open your Python script (.py file) in Visual Studio Code using the editor. You can do this by navigating to the file in the Explorer pane and clicking on it.

## 2. Install and Enable Jupyter Extension
Install the Jupyter extension from the VS Code marketplace if it is not already installed. Enable it to access notebook features such as running code cells and exporting scripts.

## 3. Convert .py Script to Jupyter Notebook Format
Use the 'Jupyter: Export as Notebook' command from the command palette (Ctrl+Shift+P) or right-click in the editor and select 'Export as Jupyter Notebook'. This will generate a .ipynb file from your .py script.

## 4. Verify and Edit Notebook Cells
Open the generated .ipynb file and verify that the code and markdown cells are correctly formatted. Edit cells as needed to improve readability or add explanations for each step.

## 5. Run and Test the Converted Notebook
Run each cell in the notebook to ensure that the code executes as expected. Use the integrated output pane and terminal in VS Code to view results and debug if necessary.
